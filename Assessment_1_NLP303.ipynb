```python
from transformers import pipeline
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-large-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-large-NER")

ner = pipeline("ner", model=model, tokenizer=tokenizer)
sentence = "Michael, a software engineer from Sydney, recently started working for the tech giant, Microsoft.\
He was thrilled to be part of an organisation that had a global presence and was excited about the opportunity\
to contribute to innovative projects. His first assignment was to travel to Paris for a collaborative project with\
the French team. Despite the challenges of relocating, Michael was eager to embark on this new journey with Microsoft in Paris."

ner_results = ner(sentence)
print(ner_results)
```

```python
sumarizer = pipeline('summarization', model = 'facebook/bart-large-cnn')
Article = """ In the early days, social media platforms like Facebook and Twitter were primarily used for connecting with friends and family. However, they have evolved into powerful tools for news dissemination, marketing, and even social activism. Today, social media is a global phenomenon that transcends geographical boundaries, making the world a smaller place.
One of the most significant impacts of social media is its role in democratizing information. Today, anyone with an internet connection can share their views with the world. This has given a voice to those who were previously unheard, leading to a more inclusive public discourse.
However, the rise of social media has also brought challenges. The spread of misinformation and fake news has become a significant concern. It has the potential to influence public opinion and even sway elections. Therefore, it’s crucial for social media platforms and users to take responsibility for verifying the information they share.
Moreover, while social media can foster connections, it can also lead to feelings of isolation and inadequacy. The constant exposure to carefully curated, idealized representations of others’ lives can lead to negative self-comparison.
In conclusion, social media is a double-edged sword. While it has the power to connect us and democratize information, it also has the potential to spread misinformation and affect our mental health. As users, it’s essential to use these platforms responsibly and maintain a critical perspective on the information we consume."""
print(sumarizer(Article, max_length=130, min_length=30, do_sample=False))
```
